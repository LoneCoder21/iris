```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
# import the necessary packages
```

```python
data = pd.read_csv("Iris.csv") # read the csv file using pandas
data.head() # show only a few tuples to get an idea of the dataset
```

```python
tuples = data.shape[0] # the rows of the table denote the number of tuples
print("a) How many tuples does the Iris data set have? -", tuples, "tuples")
```

```python
features = data[data.columns[:-1]] # ignore the last column which is the observed value
feature_columns = features.columns
print("b) How many features does the Iris data set have? -", feature_columns) # show the features
```

```python
dtypes = data.dtypes # Get the data types for the table
print("c) What’s the data type for each feature?")
print(dtypes) # show the data types
```

```python
setosa = "Iris-setosa"
versicolor = "Iris-versicolor"
virginica = "Iris-virginica"

print("d) How many of each species do you have?")

setosa_table = data.loc[data["species"] == setosa]
versicolor_table = data.loc[data["species"] == versicolor]
virginica_table = data.loc[data["species"] == virginica]
# get only the tables where the species column is of each iris type

setosa_count = setosa_table.shape[0]
versicolor_count = versicolor_table.shape[0]
virginica_count = virginica_table.shape[0]
# get the total counts of each species using the rows of the tables 

print(setosa, "-", setosa_count, "tuples")
print(versicolor, "-", versicolor_count, "tuples")
print(virginica, "-", virginica_count, "tuples")
```

```python
print("e) What’s the mean, median, quantiles and range (max-min) for each petal and sepal measurements?")
```

```python
print("Mean")
print(features.mean()) # show the mean
```

```python
print("Median")
print(features.median()) # show the median
```

```python
print("Quantile")
print(features.quantile([ .25, .50, .75])) # show the 3 quantiles at Q1 = 25%, Q2 = 50%, and Q3 = 75%
```

```python
features_max = features.max() # get the max for each feature
features_min = features.min() # get the min for each feature

features_range = features_max - features_min # range is max - min

print("Range")
print(features_range) # show the range
```

```python
print("f) Make a scatter plot of sepal_length vs sepal_width")
features.plot.scatter(x='sepal_length', y='sepal_width', c='DarkBlue', title="Sepal Width vs Sepal Length") # show the scatter plot between sepal length and sepal width
```

```python
print("g) Create four individual histogram plots in a single figure, where each plot contains one feature (petal_width, petal_length, sepal_width, sepal_length)")

features.hist() # show the 4 individual histograms using pandas
```

```python
features['petal_width'].plot.hist(alpha=0.5, label='petal_width')
features['petal_length'].plot.hist(alpha=0.5, label='petal_length')
features['sepal_width'].plot.hist(alpha=0.5, label='sepal_width')
features['sepal_length'].plot.hist(alpha=0.5, label='sepal_length')
plt.legend()
plt.title("Histogram of features")
plt.ylabel('Length and Width (centimeters)')
# showing the features histogram overlaid on top of each other
```

```python
print("h) Create four individual boxplots in a single figure, where each plot contains one feature (petal_width, petal_length, sepal_width, sepal_length)")

features.boxplot() # show the 4 individual boxplots using pandas
```

```python
print("i) Make a scatter plot matrix to examine the correlation between each of the measurements")

scatter_matrix(features, figsize=(10,10)) # show the scatter plot matrix to observe the correlations
```
